```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('tradeindia_products_structured.csv')
df.head()
```

```python
print("Rows:", len(df))
print("Columns:", len(df.columns))
print(df.info())
display(df.describe(include='all').transpose())
display(df.isnull().mean() * 100)
```

```python
# Category distribution
plt.figure(figsize=(8,4))
df['Category'].value_counts().plot(kind='bar')
plt.title('Product Counts per Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Top companies
top_companies = df['Company Name'].value_counts().head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=top_companies.values, y=top_companies.index)
plt.title('Top 10 Companies by Listings')
plt.xlabel('Listings')
plt.ylabel('Company Name')
plt.show()
```

```python
# If product name contains product type, extract some frequent words
from collections import Counter
import re

product_words = []
for x in df['Product Name'].dropna():
    product_words.extend(re.findall(r'\b\w+\b', str(x).lower()))
word_counts = Counter(product_words)
# Remove very common English words, etc.
skip_words = {'and','for','with','in','the','of','to','by','on','or','from','set','at','is'}
common = [(w,c) for w,c in word_counts.most_common(30) if w not in skip_words and len(w)>2]
print("Most frequent words in product names:", common[:10])
```

```python
# Try to extract numbers from price strings
def extract_price(s):
    if pd.isnull(s): return np.nan
    s = str(s).replace(',', '')
    for part in s.split():
        try:
            return float(part)
        except ValueError:
            continue
    return np.nan

df['PriceValue'] = df['Price'].apply(extract_price)
plt.figure(figsize=(8,4))
sns.histplot(df['PriceValue'].dropna(), bins=40)
plt.title('Price Distribution')
plt.xlabel('Price (INR)')
plt.show()
print("Price Range: ", df['PriceValue'].min(), '-', df['PriceValue'].max())
```

```python
plt.figure(figsize=(10,4))
df['Location'].value_counts().head(15).plot(kind='bar')
plt.title('Top Locations')
plt.ylabel('Supplier Count')
plt.show()

# Category-location cross
pd.crosstab(df['Category'], df['Location']).head(10)
```
